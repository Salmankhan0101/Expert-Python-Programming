## Question 1: Create a Laptop class with a private attribute __price. Implement a method set_price(new_price) to update the price and get_price() to retrieve the price.

```python
class Laptop:
    def __init__(self, price):
        self.__price = price

    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
        else:
            print("Price must be positive!")

    def get_price(self):
        return self.__price

laptop1=Laptop(2000)
print("Original Price:", laptop1.get_price())
laptop1.set_price(2500)
print("Updated Price:", laptop1.get_price())
```

### Question 2: Create a BankAccount class with a private attribute __balance. Implement methods to deposit, withdraw, and check balance while ensuring the balance cannot be modified directly.

```python
class BankAccount:
    def __init__(self,balance):
        self.__balance=balance

    def get_balance(self):
        return self.__balance

    def deposit(self,amount):
        if amount>0:
            self.__balance+=amount
            print("Amount deposit")
        else:
            print("Invalid deposit")

    def withdrow(self,amount):
        if amount<=self.__balance:
            self.__balance-=amount
            print("Withdrowal successfully")
        else:
            print("Insufficient balance")

account=BankAccount(1000)
account.deposit(500)
account.withdrow(300)
print(account.get_balance())
```

## Question 3: Create a Movie class with a private attribute __rating. Implement methods to set and get the rating, ensuring the rating is between 1 and 10.

```python
class Movie:
    def __init__(self,rating):
        self.__rating=max(1,min(10,rating))

    def set_rating(self,new_rating):
        if 1<=new_rating<=10:
            self.__rating=new_rating

    def get_rating(self):
        return self.__rating

movie1=Movie(8)
print("Initial Rating:",movie1.get_rating())

movie1.set_rating(9)
print("Updated Rating:", movie1.get_rating())

movie1.set_rating(15)
print("After Invalid Update:", movie1.get_rating())
```

## Question 4: Create a Car class with a private attribute __speed. Implement methods to increase speed, decrease speed, and get the current speed, ensuring speed does not go below 0.

```python
class Car:
    def __init__(self):
        self.__speed = 0

    def increase_speed(self, amount):
        if amount > 0:
            self.__speed += amount
        else:
            print("Increase amount must be positive.")

    def decrease_speed(self, amount):
        if amount > 0:
            self.__speed -= amount
            if self.__speed < 0:
                self.__speed = 0
        else:
            print("Decrease amount must be positive.")

    def get_speed(self):
        return self.__speed

car1 = Car()

car1.increase_speed(50)
print("Current Speed:", car1.get_speed())

car1.decrease_speed(20)
print("After Decrease:", car1.get_speed())

car1.decrease_speed(50)
print("Final Speed:", car1.get_speed())
```
