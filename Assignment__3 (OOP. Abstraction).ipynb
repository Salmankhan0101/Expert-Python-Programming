```python
from abc import ABC, abstractmethod 

class Animal(ABC):
    @abstractmethod 
    def make_sound(self):
        pass 
        
class Dog(Animal):
    def make_sound(self):  
        print("Bark")

class Cat(Animal):
    def make_sound(self): 
        print("Meow")

dog = Dog()
dog.make_sound() 
cat = Cat()
cat.make_sound()  
```

```python
from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def mileage(self):
        pass
    def color(self):
        print("white")

class Suzuki(Car):
    def mileage(self):
        print("The mileage is 25Kmph")

class Honda(Car):
    def mileage(self):
        print("The mileage is 24Kmph")

s1=Suzuki()
s1.mileage()
s1.color()

h1=Honda()
h1.mileage()
h1.color()
```

```python
class Vehicle(ABC):
    def max_speed(self):
        pass

class Car(Vehicle):
    def max_speed(self):
        return 200

class Bike(Vehicle):
    def max_speed(self):
        return 100

car=Car()
print(car.max_speed())

bike=Bike()
print(bike.max_speed())
```
