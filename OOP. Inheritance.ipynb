## Question 2: Define a Laptop class that inherits from the Computer class and includes an additional attribute, battery_life.

```python
class Computer:
    def __init__(self, brand):
        self.brand = brand

class Laptop(Computer):
    def __init__(self, brand, battery_life):
        super().__init__(brand)
        self.battery_life = battery_life

lap = Laptop("Dell", "8 hours")
print("Brand:", lap.brand)
print("Battery Life:", lap.battery_life)
```

## Question 3: Create a Smartphone class and add another child class of your choice that inherits from an ElectronicDevice class. Include an additional attribute, operating_system.

```python
class ElectronicDevice:
    def __init__(self, brand):
        self.brand = brand

class Smartphone(ElectronicDevice):
    def __init__(self, brand, operating_system):
        super().__init__(brand)
        self.operating_system = operating_system

class Tablet(ElectronicDevice):
    def __init__(self, brand, operating_system):
        super().__init__(brand)
        self.operating_system = operating_system

phone = Smartphone("Samsung", "Android")
tablet = Tablet("Apple", "iPadOS")
print(phone.brand, phone.operating_system)
print(tablet.brand, tablet.operating_system)
```

## Question 4: Implement a Musician class and create another child class of your choice that extends an Artist class. Include an additional attribute, instrument_played.

```python
class Artist:
    def __init__(self, name):
        self.name = name

class Musician(Artist):
    def __init__(self, name, instrument_played):
        super().__init__(name)
        self.instrument_played = instrument_played

class Painter(Artist):
    def __init__(self, name, style):
        super().__init__(name)
        self.style = style

musician = Musician("Arijit", "Guitar")
painter = Painter("Ravi", "Modern Art")
print(musician.name, musician.instrument_played)
print(painter.name, painter.style)
```

## Question 5: Define a FictionBook class and add another child class of your choice that inherits from the Book class. Include an additional attribute, genre.

```python
class Book:
    def __init__(self, title):
        self.title = title

class FictionBook(Book):
    def __init__(self, title, genre):
        super().__init__(title)
        self.genre = genre

class ScienceBook(Book):
    def __init__(self, title, field):
        super().__init__(title)
        self.field = field

f_book = FictionBook("Harry Potter", "Fantasy")
s_book = ScienceBook("Physics Basics", "Physics")
print(f_book.title, f_book.genre)
print(s_book.title, s_book.field)
```

## Question 6: Create a student class that takes the name & marks of 3 subjects as arguments in the constructor. Then create a method to print the average.

```python
class Student:
    def __init__(self, name, mark1, mark2, mark3):
        self.name = name
        self.mark1 = mark1
        self.mark2 = mark2
        self.mark3 = mark3

    def average(self):
        avg = (self.mark1 + self.mark2 + self.mark3) / 3
        print(f"{self.name}'s Average Marks:", avg)

student1 = Student("Salman", 80, 75, 90)
student1.average()
```
