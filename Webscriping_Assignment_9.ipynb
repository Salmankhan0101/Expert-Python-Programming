```python
## 1-Basic web scaping without header
```

```python
url="https://www.scrapethissite.com/pages/simple/"
response=requests.get(url)
soup=BeautifulSoup(response.text,"html.parser")

countries=soup.find_all("div",class_="country")

for country in countries:
  name=country.find("h3",class_="country-name").text.strip()
  capital=country.find("span",class_="country-capital").text.strip()
  population=country.find("span",class_="country-population").text.strip()
  area=country.find("span",class_="country-area").text.strip()

  print("country:",name)
  print("capital:",capital)
  print("population:",population)
  print("area:",area)
  print("++++++++++++++++++++++++++++++++++")
```

```python
## 2-Using def function,strip,get_text,text
```

```python
def get_data():
  url="https://www.scrapethissite.com/pages/simple/"
  response=requests.get(url)
  soup=BeautifulSoup(response.text,"html.parser")

  countries=soup.find_all("div",class_="country")

  for counrty in countries:
    name=country.find("h3",class_="country-name").text.strip()
    capital=country.find("span",class_="country-capital").get_text().strip()
    population=country.find("span",class_="country-population").text.strip()
    area=country.find("span",class_="country-area").get_text().strip()

    print("country:",name)
    print("capital:",capital)
    print("population:",population)
    print("area:",area)
    print("_____________________________")
get_data()
```

```python
## 3-Web scraping with header {"User-Agent": "Mozilla/5.0")
```

```python
url="https://www.scrapethissite.com/pages/simple/"
headers = {"User-Agent": "Mozilla/5.0"}
response=requests.get(url,headers=headers)
soup=BeautifulSoup(response.text,"html.parser")

countries=soup.find_all("div",class_="country")

for country in countries:
  name=country.find("h3",class_="country-name").text.strip()
  capital=country.find("span",class_="country-capital").get_text().strip()
  population=country.find("span",class_="country-population").text.strip()
  area=country.find("span",class_="country-area").get_text().strip()

  print("country:",name)
  print("capital:",capital)
  print("population:",population)
  print("area:",area)
  print("********************************")
```

```python
## 4-With enumerate
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

countries=soup.find_all("div", class_="country")

for i, country in enumerate(countries, start=1):
    name = country.find("h3",class_="country-name").text.strip()
    capital = country.find("span", class_="country-capital").text.strip()
    population = country.find("span", class_="country-population").text.strip()
    area = country.find("span", class_="country-area").text.strip()

    print(f"Country {i}:")
    print("Name:", name)
    print("Capital:", capital)
    print("Population:", population)
    print("Area:", area)
    print("///////////////////////////////////////")
```

```python
## 5- Web scraping on table and table operations
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

countries = soup.find_all("div", class_="country")

for country in countries:
    name = country.find("h3",class_="country-name").text.strip()
    population = country.find("span", class_="country-population").text.strip()

    print(name, population)
print("Total countries:", len(countries))
```
