# 1.Write a Python program to demonstrate polymorphism where two different classes have the same method name, and a common function calls those methods.

```python

# Defining the first class
class Dog:
    def speak(self):
        return "Woof!"

# Defining the second class
class Cat:
    def speak(self):
        return "Meow!"

# Common function that calls the speak method
def animal_sound(animal):
    print(animal.speak())

# Creating instances of both classes
dog = Dog()
cat = Cat()

# Calling the common function with different class instances
animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!
```

# 2. Create a base class Vehicle with a method speed(). Then, create a derived class Car that overrides the speed() method.

```python
# 2. Create a base class Vehicle with a method speed(). Then, create a derived class Car that overrides the speed() method.
# Base class Vehicle
class Vehicle:
    def speed(self):
        print("The vehicle moves at a standard speed.")

# Derived class Car that overrides the speed method
class Car(Vehicle):
    def speed(self):
        print("The car moves at a speed of 120 km/h.")

# Creating objects and calling speed() method
vehicle = Vehicle()
vehicle.speed()  # This will call the speed method of Vehicle

car = Car()
car.speed()  # This will call the overridden speed method of Car
```

# 3.Create a class Box that overloads the + operator to add the volume of two boxes.

```python
# 3.Create a class Box that overloads the + operator to add the volume of two boxes.
class Box:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height

    def volume(self):
        return self.length * self.width * self.height

    # Overloading the + operator to add volumes of two boxes
    def __add__(self, other):
        if isinstance(other, Box):
            return self.volume() + other.volume()
        return NotImplemented

# Example usage
box1 = Box(3, 4, 5)
box2 = Box(2, 3, 6)

total_volume = box1 + box2  # This uses the overloaded + operator
print("Total Volume:", total_volume)
```
